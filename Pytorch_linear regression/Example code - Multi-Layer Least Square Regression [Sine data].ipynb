# Least Square Regression - Multi-Layer Case
## (for sine wave data)

# Library

```python
import numpy as np

import torch
import torch.nn as nn # for neural network models
import torch.optim as optim # for optimization 
import torch.nn.init as init # for initialization 

from matplotlib import pyplot as plt
```

# simple data

```python
# true data
N = 100
X = np.linspace(-np.pi, np.pi, N).reshape(N, 1)
real_y = np.sin(X)

# add noise
y = real_y + 0.1*np.random.randn(N, 1)

# shuffle
idx = np.random.permutation(N)
X_ = X[idx]
y_ = y[idx]
```

```python
# show
plt.figure()
plt.plot(X, real_y)
plt.scatter(X, y)
plt.show()
```

# Split train / test dataset

```python
n = N-20

X_train = X_[:n]
y_train = y_[:n]
X_test = X_[n:]
y_test = y_[n:]

```

```python
plt.figure()
plt.scatter(X_train, y_train, label='train')
plt.scatter(X_test, y_test, label='test')
plt.legend()
plt.show()
```

# Multi-layer Linear Regression

## Hyper parameter setting

```python
learning_rate = 1e-1
epochs = 1000
```

## Make model with pytorch (Gradient Descent)

```python
# prepare data
X_train = torch.tensor(X_train, dtype = torch.float) # if X_train is numpy.array
label = torch.tensor(y_train, dtype = torch.float)

# prepare model
model = nn.Sequential(
        nn.Linear(1,1),
        nn.Sigmoid(),
        nn.Linear(1,1),
        )

# prepare optimization and loss function
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []
# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward()
    optimizer.step()
    
    loss_array.append(loss)
    
X_test = torch.tensor(X_test, dtype = torch.float)
X_ = torch.tensor(X_, dtype = torch.float)

y_train_pred = model(X_train)
y_whole_pred = model(X_)

y_train_pred = y_train_pred.detach().numpy()
y_whole_pred = y_whole_pred.detach().numpy()
```

```python
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.legend()
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.legend()
plt.title('whole data')
plt.show()
```

```python
### plot loss
plt.figure()
plt.plot(loss_array)
plt.show()
            
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```

# Sum-up

```python
#################### hyper parameter setting ####################
learning_rate=1e-1
epochs = 1000

h = 10 ### number of hidden layer nodes
act = nn.Tanh() ### activation function
#################################################################
# prepare data
X_train = torch.tensor(X_train, dtype = torch.float)
label = torch.tensor(y_train, dtype = torch.float)

# prepare model
model = nn.Sequential(
        nn.Linear(1,h),
        act,
        nn.Linear(h,200)
        act
        nn.Linear(200,1),
        )

# prepare optimziation and loss function 
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.1)

loss_array = []

# train the model
for i in range(epochs):
    optimizer.zero_grad()
    output = model(X_train)
    
    loss = loss_function(output, label)
    loss.backward() #gradient 해주기
    optimizer.step()
    
    loss_array.append(loss)
    
X_test = torch.tensor(X_test, dtype = torch.float)
X_ = torch.tensor(X_, dtype = torch.float)

y_train_pred = model(X_train)
y_whole_pred = model(X_)

y_train_pred = y_train_pred.detach().numpy()
y_whole_pred = y_whole_pred.detach().numpy()
```

```python
### plot loss
plt.figure()
plt.plot(loss_array)
plt.show()
            
plt.figure()
plt.scatter(X_train, y_train_pred, label='pred')
plt.scatter(X_train, y_train, label='target')
plt.title('train data')
plt.show()

plt.figure()
plt.scatter(X_, y_whole_pred, label='pred')
plt.scatter(X_, y_, label='target')
plt.title('whole data')
plt.show()
```
